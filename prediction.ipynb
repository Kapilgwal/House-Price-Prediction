```python
df = df.rename(columns=  {"Children'splayarea" : "childrensplayarea"})
```

```python
def applyFilter(num):
    if num == 9:
        num = 1
    return num
```

```python
df['Resale'] = df['Resale'].apply(applyFilter)
df['MaintenanceStaff'] = df['MaintenanceStaff'].apply(applyFilter)
df['Gymnasium'] = df['Gymnasium'].apply(applyFilter)
df['SwimmingPool'] = df['SwimmingPool'].apply(applyFilter)
df['LandscapedGardens'] = df['LandscapedGardens'].apply(applyFilter)
df['JoggingTrack'] = df['JoggingTrack'].apply(applyFilter)
df['RainWaterHarvesting'] = df['RainWaterHarvesting'].apply(applyFilter)
df['IndoorGames'] = df['IndoorGames'].apply(applyFilter)
df['ShoppingMall'] = df['ShoppingMall'].apply(applyFilter)
df['Intercom'] = df['Intercom'].apply(applyFilter)
df['SportsFacility'] = df['SportsFacility'].apply(applyFilter)
df['ATM'] = df['ATM'].apply(applyFilter)
df['ClubHouse'] = df['ClubHouse'].apply(applyFilter)
df['School'] = df['School'].apply(applyFilter)
df['24X7Security'] = df['24X7Security'].apply(applyFilter)
df['PowerBackup'] = df['PowerBackup'].apply(applyFilter)
df['CarParking'] = df['CarParking'].apply(applyFilter)
df['StaffQuarter'] = df['StaffQuarter'].apply(applyFilter)
df['Cafeteria'] = df['Cafeteria'].apply(applyFilter)
df['MultipurposeRoom'] = df['MultipurposeRoom'].apply(applyFilter)
df['Hospital'] = df['Hospital'].apply(applyFilter)
df['WashingMachine'] = df['WashingMachine'].apply(applyFilter)
df['Gasconnection'] = df['Gasconnection'].apply(applyFilter)
df['AC'] = df['AC'].apply(applyFilter)
df['Wifi'] = df['Wifi'].apply(applyFilter)
df['childrensplayarea'] = df['childrensplayarea'].apply(applyFilter)
df['LiftAvailable'] = df['LiftAvailable'].apply(applyFilter)
df['BED'] = df['BED'].apply(applyFilter)
df['VaastuCompliant'] = df['VaastuCompliant'].apply(applyFilter)
df['Microwave'] = df['Microwave'].apply(applyFilter)
df['GolfCourse'] = df['GolfCourse'].apply(applyFilter)
df['TV'] = df['TV'].apply(applyFilter)
df['DiningTable'] = df['DiningTable'].apply(applyFilter)
df['Wardrobe'] = df['Wardrobe'].apply(applyFilter)
df['Sofa'] = df['Sofa'].apply(applyFilter)
df['Refrigerator'] = df['Refrigerator'].apply(applyFilter)
```
