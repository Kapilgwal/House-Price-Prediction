# Applying algorithms

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['city','Resale','No. of Bedrooms','24X7Security','PowerBackup','CarParking','LiftAvailable']]
Y = df[['Price']]
```

```python
def applyFilter(text):
    if text == 'kolkata':
      text = 1
    if text == 'mumbai':
      text = 2
    if text == 'delhi':
      text = 3
    if text == 'chennai':
      text = 4
    if text == 'bangalore':
      text = 5
    if text == 'hyderabad':
      text = 6
    return text
```

```python
X['city'] = X['city'].apply(applyFilter)
```

```python
X
```

```python
Y
```

```python
X_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred = y_pred.astype(int)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rdc = RandomForestClassifier(n_estimators=100,random_state=42)
rdc.fit(X_train,y_train)
y_pred = rdc.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy*100
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
gb = GaussianNB()
gb.fit(X_train,y_train)
y_pred = gb.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import accuracy_score
lcv = LassoCV()
lcv.fit(X_train,y_train)
y_pred = lcv.predict(x_test).astype(int)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import RidgeClassifierCV
from sklearn.metrics import accuracy_score
lcv = RidgeClassifierCV()
lcv.fit(X_train,y_train)
y_pred = lcv.predict(x_test).astype(int)
accuracy_score(y_pred,y_test)
```
